```python
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B','A', 'B', 'C', 'D'],
    'Value': [10, 20, 15, 25, 10, 30, 50, 45]
}

df = pd.DataFrame(data)
df['Value']
```

```python
df['Value'].agg(sum) #select sum(value) from df
```

```python
df['Value'].agg(len) #select len(value) from df
```

```python
df['Value'].agg(lambda x: x**5) #select x**5(value) from df
```

```python
grouped = df.groupby(['Category'])
print(type(grouped))

# when grouped it is going to create smaller groups
```

```python
grouped.ngroups
```

```python
for i in grouped:
    print(i)

for i in grouped:
    print(type(i))
```

```python
# several agg in 1 

grouped['Value'].agg(['sum', 'min', 'max', 'mean'])
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'], 
    'Subcategory': ['X', 'X', 'Y', 'Y', 'x', 'Y'], 
    'Value': [10, 20, 15, 25, 10, 30]
}

df = pd.DataFrame(data) 

df
```

```python
df.groupby(['Category', 'Subcategory'])['Value'].agg(['sum', 'min', 'max', 'mean'])
```

```python
df.groupby(['Category'])['Value'].agg(['sum', 'min', 'max', 'mean'])
# select category, subcategory sum(value), avg(value), max(value), min (value) from df group by category, subcategory
```

```python
# concat -- from pandas itself, vertical & horizontal
# join -- from dataframe, horizontal only
# merge -- from dataframe, horizontal only
```

```python
df_16 = pd.read_csv('stocks_16.csv')
df_17 = pd.read_csv('stocks_17.csv')
df_16
```

```python
df_17
```

```python
df_16.concat()
```

```python
# concat default - vertical, full outer join
# join default - left join
# merge - inner join

dfs = pd.concat([df_16, df_17])
dfs
```

```python
dfs = pd.concat([df_16, df_17], axis=1)
dfs
```

```python
# joins the indexes - possible to change when reading the file

dfs = pd.concat([df_16, df_17], axis=1, join='inner')
dfs
```

```python
df_16 = pd.read_csv('stocks_16.csv', index_col='Symbol')
df_17 = pd.read_csv('stocks_17.csv', index_col='Symbol')
```

```python
dfs = pd.concat([df_16, df_17], axis=1, join='inner')
dfs
```

```python
# keys - define from which col are coming

dfs = pd.concat([df_16, df_17], axis=1, keys=['2016', '2017'], join='outer')
dfs
```

```python
# ambisious, empID
# solution - give lsuffix

dfs2 = df_16.join(df_17)
```

```python
dfs2 = df_16.join(df_17, lsuffix='_2016', rsuffix='_2017')
dfs2
```

```python
dfs2 = df_16.join(df_17, lsuffix='_2016', rsuffix='_2017',how='outer')
dfs2
```

```python
# instead of r and lsuffix can use add_suffix

# in both join & concat it joins looking at the col

dfs2 = df_16.add_suffix('_2016').join(df_17.add_suffix('_2017'),how='right')
dfs2
```

```python
# in merge - column should be given

dfs3 = df_16.merge(df_17, on='Symbol', how='inner')
dfs3
```

```python
dfs3 = df_16.merge(df_17, on=['Symbol', ' Shares'], how='outer')
dfs3
```

```python
# Data for df1
data1 = {'ProductID': [1, 2, 3],
         'ProductName': ['Widget', 'Gadget', 'Doodad'],
         'Category': ['Tools', 'Electronics', 'Toys']}

df1 = pd.DataFrame(data1)

# Data for df2
data2 = {'ProductID': [1, 2, 4],
         'Price': [19.99, 29.99, 15.99],
         'Stock': [100, 200, 150]}

df2 = pd.DataFrame(data2)
merged_df = df1.merge(df2, on='ProductID', how='inner')

print(merged_df)
```

```python
# Data for customers
data_customers = {'CustomerID': [1, 2, 3],
                  'CustomerName': ['Alice', 'Bob', 'Charlie'],
                  'Country': ['USA', 'UK', 'Canada']}

customers = pd.DataFrame(data_customers)

# Data for orders
data_orders = {'OrderID': [101, 102, 103],
               'CustomerID': [1, 1, 3],
               'OrderDate': ['2025-01-15', '2025-02-15', '2025-03-15']}

orders = pd.DataFrame(data_orders)

dfs = customers.merge(orders,on='CustomerID',how='left')
dfs
```

```python
# Data for sales_jan
data_jan = {'Date': ['2025-01-01', '2025-01-02', '2025-01-03'],
            'ProductID': [1, 2, 3],
            'Sales': [10, 20, 30]}

sales_jan = pd.DataFrame(data_jan)

# Data for sales_feb
data_feb = {'Date': ['2025-02-01', '2025-02-02', '2025-02-03'],
            'ProductID': [1, 2, 3],
            'Sales': [15, 25, 35]}

sales_feb = pd.DataFrame(data_feb)

# Data for sales_mar
data_mar = {'Date': ['2025-03-01', '2025-03-02', '2025-03-03'],
            'ProductID': [1, 2, 3],
            'Sales': [20, 30, 40]}

sales_mar = pd.DataFrame(data_mar)



all_sales = pd.concat([sales_mar, sales_jan, sales_feb], ignore_index=True).sort_values(by='Date')
all_sales
```

```python
# Data for df1
data_emp1 = {'EmployeeID': [1, 2, 3],
             'Name': ['John', 'Jane', 'Doe'],
             'Department': ['HR', 'Finance', 'IT']}

df1 = pd.DataFrame(data_emp1)

# Data for df2
data_emp2 = {'EmployeeID': [1, 3, 4],
             'Salary': [50000, 60000, 55000],
             'JoinDate': ['2023-01-10', '2024-02-15', '2025-03-01']}

df2 = pd.DataFrame(data_emp2)


dfs = df1.merge(df2, on='EmployeeID', how='left')
dfs

```

```python
# Data for students
data_students = {'StudentID': [1, 2, 3],
                 'Name': ['Anna', 'Bill', 'Cathy'],
                 'Major': ['Math', 'Physics', 'Chemistry']}

students = pd.DataFrame(data_students)

# Data for courses
data_courses = {'CourseID': [101, 102, 103],
                'CourseName': ['Calculus', 'Mechanics', 'Organic Chemistry'],
                'Credits': [4, 3, 4]}

courses = pd.DataFrame(data_courses)

# Data for enrollments
data_enrollments = {'StudentID': [1, 1, 2],
                    'CourseID': [101, 102, 103],
                    'Grade': ['A', 'B+', 'A-']}

enrollments = pd.DataFrame(data_enrollments)


student_enrollments = students.merge(enrollments, on='StudentID', how='inner')

all_info = student_enrollments.merge(courses, on='CourseID' , how='inner')
all_info
```
