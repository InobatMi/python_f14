```python
# Homework 2:
import pandas as pd
stackoverflow_df = pd.read_csv('task\\stackoverflow_qa.csv')
stackoverflow_df.head()

# Convert 'creationdate' to datetime format
stackoverflow_df['creationdate'] = pd.to_datetime(stackoverflow_df['creationdate'])

# 1. Find all questions that were created before 2014
print(stackoverflow_df[stackoverflow_df['creationdate'].dt.year < 2014])

# 2. Find all questions with a score more than 50
print(stackoverflow_df[stackoverflow_df['score'] > 50])

# 3. Find all questions with a score between 50 and 100
print(stackoverflow_df[stackoverflow_df['score'].between(50, 100)])

# 4. Find all questions answered by Scott Boston
print(stackoverflow_df[stackoverflow_df['ans_name'] == 'Scott Boston'])

# 5. Find all questions answered by the following 5 users
print(stackoverflow_df[stackoverflow_df['ans_name'].isin(['emesday', 'Jeff', 'Chang She', 'dukebody', 'FooBar'])])

# 6. Find all questions that were created between March 2014 and 
# October 2014 that were answered by Unutbu and have a score less than 5.
filter26_1 = stackoverflow_df['creationdate'].between('2014-03-01', '2014-10-31')
filter26_2 = stackoverflow_df['ans_name'] == 'unutbu'
filter26_3 = stackoverflow_df['score'] < 5
print(stackoverflow_df[filter26_1 & filter26_2 & filter26_3])

# 7. Find all questions that have a score between 5 and 10 or have a view count greater than 10,000
filter27_1 = stackoverflow_df['score'].between(5,10)
filter27_2 = stackoverflow_df['viewcount'] > 10000
print(stackoverflow_df[filter27_1 | filter27_2])

# 8. Find all questions that are not answered by Scott Boston
print(stackoverflow_df[~(stackoverflow_df['ans_name'] == 'Scott Boston')])
```

```python
# Homework 3:

# Titanic data set, stored as CSV. The data consists of the following data columns:

# - PassengerId: Id of every passenger.
# - Survived: Indication whether passenger survived. 0 for yes and 1 for no.
# - Pclass: One out of the 3 ticket classes: Class 1, Class 2 and Class 3.
# - Name: Name of passenger.
# - Sex: Gender of passenger.
# - Age: Age of passenger in years.
# - SibSp: Number of siblings or spouses aboard.
# - Parch: Number of parents or children aboard.
# - Ticket: Ticket number of passenger.
# - Fare: Indicating the fare.
# - Cabin: Cabin number of passenger.
# - Embarked: Port of embarkation.


titanic_df = pd.read_csv("task\\titanic.csv")


# | PassengerId | Survived | Pclass | Name                                             | Sex    | Age   | SibSp | Parch | Ticket           | Fare    | Cabin | Embarked |
# |-------------|----------|--------|--------------------------------------------------|--------|-------|-------|-------|------------------|---------|-------|----------|
# | 1           | 0        | 3      | Braund, Mr. Owen Harris                         | male   | 22.0  | 1     | 0     | A/5 21171        | 7.2500  | NaN   | S        |
# | 2           | 1        | 1      | Cumings, Mrs. John Bradley (Florence Briggs Th.) | female | 38.0  | 1     | 0     | PC 17599         | 71.2833 | C85   | C        |
# | 3           | 1        | 3      | Heikkinen, Miss. Laina                          | female | 26.0  | 0     | 0     | STON/O2. 3101282 | 7.9250  | NaN   | S        |
# | 4           | 1        | 1      | Futrelle, Mrs. Jacques Heath (Lily May Peel)    | female | 35.0  | 1     | 0     | 113803           | 53.1000 | C123  | S        |
# | 5           | 0        | 3      | Allen, Mr. William Henry                        | male   | 35.0  | 0     | 0     | 373450           | 8.0500  | NaN   | S        |


# 1. Select Female Passengers in Class 1 with Ages between 20 and 30:
# Extract a DataFrame containing female passengers in Class 1 with ages between 20 and 30.

filter31_1 = titanic_df['Sex'] == 'female'
filter31_2 = titanic_df['Pclass'] == 1
filter31_3 = titanic_df['Age'].between(20,30)
print(titanic_df[filter31_1 & filter31_2 & filter31_3])



# 2. Filter Passengers Who Paid More than $100:
# Create a DataFrame with passengers who paid a fare greater than $100.
high_fare_passengers = titanic_df[titanic_df['Fare'] > 100]
print(high_fare_passengers)

# 3. Select Passengers Who Survived and Were Alone:
# Filter passengers who survived and were traveling alone (no siblings, spouses, parents, or children).
filter33_1 = titanic_df['SibSp'] == 0 
filter33_2 = titanic_df['Parch'] == 0 
filter33_3 = titanic_df['Survived'] == 0 
print(titanic_df[filter33_1 & filter33_2 & filter33_3])


# 4. Filter Passengers Embarked from 'C' and Paid More Than $50:
# Create a DataFrame with passengers who embarked from 'C' and paid more than $50.

filter34_1 = titanic_df['Embarked'] == 'C'
filter34_2 = titanic_df['Fare'] > 50
c_high_fare_pass = titanic_df[filter34_1 & filter34_2]
print(c_high_fare_pass)

# 5. Select Passengers with Siblings or Spouses and Parents or Children:
# Extract passengers who had both siblings or spouses aboard and parents or children aboard.

filter35_1 = titanic_df['Parch'] > 0
filter35_2 = titanic_df['SibSp'] > 0
print(titanic_df[filter35_1 & filter35_2])

# 6. Filter Passengers Aged 15 or Younger Who Didn't Survive:
# Create a DataFrame with passengers aged 15 or younger who did not survive.


filter36_1 = titanic_df['Age'] <= 15
filter36_2 = titanic_df['Survived'] == 1
young_died_passengers = titanic_df[filter36_1 & filter36_2]
print(young_died_passengers)

# 7. Select Passengers with Cabins and Fare Greater Than $200:
# Extract passengers with known cabin numbers and a fare greater than $200.

filter37_1 = titanic_df['Cabin'].notna()
filter37_2 = titanic_df['Fare'] > 200
print(titanic_df[filter37_1 & filter37_2])

# 8. Filter Passengers with Odd-Numbered Passenger IDs:
# Create a DataFrame with passengers whose PassengerId is an odd number.

odd_passengerids = titanic_df[titanic_df['PassengerId'] % 2 == 1]
print(odd_passengerids)

# 9. Select Passengers with Unique Ticket Numbers:
# Extract a DataFrame with passengers having unique ticket numbers.

unique_tickets = titanic_df['Ticket'].value_counts() 
unique_passengers_df = titanic_df[titanic_df['Ticket'].isin(unique_tickets[unique_tickets == 1].index)]
print(unique_passengers_df)

# 10. Filter Passengers with 'Miss' in Their Name and Were in Class 1:
# Create a DataFrame with female passengers having 'Miss' in their name and were in Class 1.

filter310_1 = titanic_df['Name'].str.contains(r'\bMiss\b', case=False, na=False)
filter310_2 = titanic_df['Sex'] == 'female'
filter310_3 = titanic_df['Pclass'] == 1
print(titanic_df[filter310_1 & filter310_2 & filter310_3])
```
