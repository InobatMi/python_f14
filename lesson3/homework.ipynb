```python
# Homework2:

# Create a list named food with two elements, "rice" and "beans".

products_list = ["rice", "beans"]

# Append the string "broccoli" to food using .append().

products_list.append("broccoli")

# Add the strings "bread" and "pizza" to food using .extend().

products_list.extend(["bread", "pizza"])

# Print the first two items in the food list using print() and slice notation.

print(products_list[:2])

# Print the last item in food using print() and index notation.

print(products_list[-1])

# Create a list called breakfast from the string "eggs, fruit, orange, juice" using the string .split() method. 

breakfast = "eggs, fruit, orange, juice".split(', ')
print(breakfast)

# 7. Verify that breakfast has three items using len(). 

print(breakfast.__len__())

# 8. Create a new list called lengths using a list comprehension that contains the lengths of each string in 
# the breakfast list


lengths = [len(i) for i in breakfast]

lengths

```
