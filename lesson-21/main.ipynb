```python
import json
import os
import requests

with open('politian.json', encoding='utf-8') as f:
    data = json.load(f)
    
for politian in data:
    os.mkdir(politian['name'])
    df = pd.DataFrame([politian])
    
    filtered = df[['name', 'dob', 'country','party','position','years_in_office']]
    filtered.to_csv(f'{politian["name"]}/{politian["name"]}.csv', index=False)
    
    image = politian['image']
    respond = requests.get(image)
        
    with open(os.path.join(f'{politian["name"]}', f"{politian["name"]}.jpg"), 'wb') as f:
        f.write(respond.content)
    
    biography = df['biography']
    biography.to_csv(f'{politian["name"]}/{politian["name"]}', index=False)
    
```
